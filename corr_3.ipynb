```python
from generate_configurations import read_configurations
import numpy as np
import matplotlib.pyplot as plt
from corr_3 import *
from ising import exact_heat_capacity
```

```python
configurations , _, _, params = read_configurations('data/1000_2.npz')
```

```python
rs = np.linspace(0, 1000, 100)
cr = []
```

```python
cr = np.array([corr_3(configurations, r) for r in rs])
```

```python
cr_an = np.array([exact_heat_capacity(params['beta'], params['J'], params['size'], r) for r in rs])
```

```python
plt.plot(rs, cr, label='Simulation', marker='x')
plt.plot(rs, cr_an, label='Analytical', marker='+')
plt.xlabel('r')
plt.ylabel('C(r)')
plt.title('Correlation function')
plt.legend()
```
