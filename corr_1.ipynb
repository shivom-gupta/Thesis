```python
from generate_configurations import read_configurations
import numpy as np
import matplotlib.pyplot as plt
from corr_1 import *
from ising import exact_heat_capacity
```

```python
configurations , _, _, params = read_configurations('data/100_0.5.npz')
```

```python
rs = np.linspace(0, 10, 11)
cr = []
```

```python
cr = np.array([cr_1(configurations, r) for r in rs])
```

```python
cr_op = np.array([cr_1_optimized(configurations, r) for r in rs])
```

```python
cr_an = np.array([exact_heat_capacity(params['beta'], params['J'], params['size'], r) for r in rs])
```

```python
rs_off = rs +0.2

plt.scatter(rs, cr_op, label='Simulation', marker='x')
plt.errorbar(rs, cr_op, yerr = abs(cr_op - cr_an), label='Error')
plt.plot(rs_off, cr_an, label='Analytical', marker='+')
plt.xlabel('r')
plt.ylabel('C(r)')
plt.title('Correlation function (with offset in r)')
plt.legend()
```
